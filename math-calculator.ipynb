**Example of plotting a function**

```python
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 2*np.power(t, 3)-1.5*np.power(t, 2)-t

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='some function',
       title='Your title')
ax.grid()

plt.show()
```

**Find the zeros of a function using SymPy**

In this case, the function is $f\left( x \right) = 3x^3+5.2x^2-4.2x-5.3$

```python
from __future__ import division
x, y, z, t = sympy.symbols('x y z t')
k, m, n = sympy.symbols('k m n', integer=True)
f, g, h = sympy.symbols('f g h', cls=sympy.Function)
f = 3*x**3+5.3*x**2-4.2*x-5.3
sympy.solveset(f, x, domain=sympy.S.Reals)
```
